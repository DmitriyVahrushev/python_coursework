```python
def A(m, n):
    if m == 0:
        return n + 1
    elif m > 0 and n == 0:
        return A(m-1, 1)
    elif m > 0 and n > 0:
        k = A(m, n - 1)
        return A(m - 1, k)
```

```python
A(4, 4)
```

```python
passes = []
with open('passwords.txt') as file:
    for line in file:
        passes.append(line.rstrip('\n') + " wrong")
```

```python
with open('results.docx', 'w') as file:
    for password in passes:
        file.write(password + '\n')
```
